```python
from ultralytics import YOLO
```

* Inference with YOLO11 model

```python
model = YOLO('yolo11n.pt')

results = model.predict("/content/football.mp4", save=True)
print(results[0])

print("#################################################")
print("\n")


for box in results[0].boxes:
  print(box)
```

```python
# Get bounding box, confidence score and data as Tensors

for result in results:
  print("bounding box :", result.boxes.xywh)
  print("\n")
  print("confidence score :", result.boxes.conf)
  print("\n")
  print("data :", result.boxes.data)
  print("#################################################")
```
